**Import Libraries and Dataset**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titles_df = pd.read_csv('Movie_Id_Titles')
ratings_df = pd.read_csv('u.data',sep='\t',names=['user_id','item_id','rating','timestamp'])
```

```python
ratings_df
```

```python
ratings_df.drop(['timestamp'],axis=1,inplace=True)
```

```python
df=pd.merge(ratings_df,titles_df,on='item_id')
```

**Visualizing the Dataset**

```python
df
```

```python
df.groupby('title').describe()
```

```python
df.groupby('title')['rating'].describe()
```

```python
mean_df = df.groupby('title')['rating'].describe()['mean']
```

```python
count_df = df.groupby('title')['rating'].describe()['count']
```

```python
mean_df
```

```python
count_df
```

```python
df_rating =pd.concat([count_df,mean_df],axis=1)
```

```python
df_rating
```

```python
df_rating.reset_index()
```

```python
df_rating['mean'].plot(bins=100, kind='hist',color='red')
```
